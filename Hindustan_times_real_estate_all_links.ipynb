```python
import requests
```

```python
import bs4
import json
```

```python
res=requests.get("https://www.hindustantimes.com/real-estate/")
```

```python
soup=bs4.BeautifulSoup(res.text,"lxml")
```

```python
type(soup)                        
```

```python
title=soup.select("title")
```

```python
title
```

```python
l=[]
for link in soup.select("a",href=True):
    if ("https://www.hindustantimes.com/real-estate" in link['href']):
        l.append(link['href'])
        print(link['href'])
```

```python
#s=json.dumps(l)
#with open("links.json","w") as f:
#    json.dump(s,f)
    
    
l
```

```python
data={}
sed={}
i=1
for li in l:
    article=requests.get(li)
    cont=article.content
    doc=bs4.BeautifulSoup(cont,"html.parser")
    Heading=doc.find("h1")
    title=Heading.text
    Division=(doc.find_all("div",{"class": "storyDetail"}))
    docu=""
    for lw in Division:
        par=lw.find_all("p")
        print(title)
        for t in par:
            print(t.text)
            docu=''.join([docu,t.text])
    sed["Title"]=title
    sed["Article"]=docu  
    data[i]=sed
    i+=1
```

```python
data
```

```python
df=json.dumps(data,indent=4,default=str)
with open("scrape.json","w") as d:
    json.dump(df,d)
```
